```python
import numpy as np

# Define valid values
valid_values = ['low', 'medium', 'high']

# Replace values in 'attacking_work_rate' that are not in valid_values with NaN
data_frames['players.csv']['attacking_work_rate'] = data_frames['players.csv']['attacking_work_rate'].apply(
    lambda x: x if x in valid_values else np.nan
)
```

```python
data_frames['players.csv']['defensive_work_rate_numeric'] = pd.to_numeric(data_frames['players.csv']['defensive_work_rate'], errors='coerce')
```

```python
# Define a function to categorize defensive_work_rate_numeric
def categorize_defensive_work_rate(value):
    if 0 <= value <= 2:
        return 'low'
    elif 3 <= value <= 6:
        return 'medium'
    elif 7 <= value <= 9:
        return 'high'
    else:
        return np.nan  # Handle unexpected values

# Apply the function to create a new column
data_frames['players.csv']['defensive_work_rate_category'] = data_frames['players.csv']['defensive_work_rate_numeric'].apply(categorize_defensive_work_rate)
```

```python
# Define valid categories
valid_categories = ['low', 'medium', 'high']

# Replace values in 'defensive_work_rate' that are not in valid_categories with the corresponding value from 'defensive_work_rate_category'
data_frames['players.csv']['defensive_work_rate'] = data_frames['players.csv'].apply(
    lambda row: row['defensive_work_rate'] if row['defensive_work_rate'] in valid_categories else row['defensive_work_rate_category'],
    axis=1
)

# Drop the 'defensive_work_rate_numeric' and 'defensive_work_rate_category' columns
data_frames['players.csv'].drop(columns=['defensive_work_rate_numeric', 'defensive_work_rate_category'], inplace=True)
```

# Handling NA value

```python
# Fill NA by modes
modes = data_frames['players.csv'][list(players_categorical_cols)].mode()
for col in list(players_categorical_cols):
    data_frames['players.csv'][col] = data_frames['players.csv'][col].fillna(modes[col][0])
```

## Handling categorical value

```python
from sklearn.preprocessing import OrdinalEncoder

category_order = ['low', 'medium', 'high', 'PhD']
encoder = OrdinalEncoder(categories=[category_order])

data_frames['players.csv']['attacking_work_rate'] = encoder.fit_transform(data_frames['players.csv'][['attacking_work_rate']])
data_frames['players.csv']['defensive_work_rate'] = encoder.fit_transform(data_frames['players.csv'][['defensive_work_rate']])
```

```python
# Assuming players_categorical_cols is already defined in your notebook
# Perform one-hot encoding on the 'preferred_foot' column
one_hot_encoded = pd.get_dummies(data_frames['players.csv']['preferred_foot'], prefix='preferred_foot')

data_frames['players.csv'] = pd.concat([data_frames['players.csv'], one_hot_encoded], axis=1)

# # Drop the original 'preferred_foot' column
data_frames['players.csv'].drop('preferred_foot', axis=1, inplace=True)
```

Drop all na rows since NA ratio is not significant (around 15%)

```python
df_clean = data_frames['games.csv'].dropna()
print(f"NA ratio is {1 - len(df_clean) / len(data_frames['games.csv'])}")
data_frames['games.csv'].dropna(inplace=True)
```
