```python
import pandas as pd

df = pd.read_csv('data/games.csv')
```

```python
df.isna().any()
```

```python
home_formations_col = [col for col in df.columns if 'home_player_X' in col or 'home_player_Y' in col]
away_formations_col = [col for col in df.columns if 'away_player_X' in col or 'away_player_Y' in col]
df['home_formations'] = df[home_formations_col].apply(lambda col: list(col.values), axis=1)
df['away_formations'] = df[away_formations_col].apply(lambda col: list(col.values), axis=1)
```

```python
# home_formations_col
df[home_formations_col]
df[away_formations_col]
```

```python
import hashlib

# Convert the list values to a unique integer using a hashing function
def list_to_int(lst):
    # Convert list to a string
    lst_str = str(lst)
    # Create a hash object
    hash_object = hashlib.md5(lst_str.encode())
    # Convert hash to an integer
    return int(hash_object.hexdigest(), 16)

df['home_formations'] = df['home_formations'].apply(list_to_int)

# Display the DataFrame with the converted column
print(df[['home_formations']])
```
